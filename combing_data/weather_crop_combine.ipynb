```python
import pandas as pd
from dotenv import load_dotenv
from helper import unzip_gz, zip_gz

pd.set_option('display.max_columns', None)
```

```python
crop_df = unzip_gz("../data/clean_crop_27_v2.gz")
crop_df.head()
```

```python
weather_df = pd.read_csv("../data/clean_weather_data.csv")
weather_df.head()
```

```python
print("Crop Data")
display(crop_df.info())
print("Weather Data")
display(weather_df.info())
```

```python
weather_df['state'] = weather_df['state'].str.upper()
weather_df['state'].value_counts()
```

```python
crop_df['LOCATION_DESC'].value_counts()
```

```python
crop_df = crop_df[~crop_df['LOCATION_DESC'].str.contains('OTHER STATES|US TOTAL', case=False, na=False)]
crop_df['LOCATION_DESC'].value_counts()
```

```python
crop_df = crop_df[(crop_df['FREQ_DESC'] == 'ANNUAL') & (crop_df['REFERENCE_PERIOD_DESC'] == 'YEAR') & (crop_df['DOMAINCAT_DESC'] == "NOT SPECIFIED")]
```

```python
crop_df
```

```python
crop_df['COMMODITY_DESC'].value_counts()
```

```python
crop_df = crop_df.drop(columns=['AGG_LEVEL_DESC', 'ASD_DESC', 'COUNTY_NAME', 'WEEK_ENDING',
                                'FREQ_DESC', 'REFERENCE_PERIOD_DESC', 'DOMAINCAT_DESC',
                                'PRODN_PRACTICE_DESC', 'CLASS_DESC', 'DOMAIN_DESC'])
crop_df
```

```python
merged_df = pd.merge(
    crop_df,
    weather_df,
    left_on=['LOCATION_DESC', 'YEAR'],
    right_on=['state', 'year'],
    how='right'
)

merged_df
```
