```python
location_list = filtered_df['LOCATION_DESC'].value_counts()

all_states = [
    "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida",
    "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine",
    "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska",
    "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota",
    "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee",
    "Texas", "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"
]
states = []

for key, value in location_list.items():
    if value >= 30000:
        if key == 'US TOTAL':
            continue
        else:
            states.append(key.title())
states.sort()


all_states = set([state.lower() for state in all_states])
state_set = set([state.lower() for state in states])

missing_states = all_states - state_set

# If we don't get an output for missing_states then we are not missing any states
missing_states
```

##### Alaska Hawaii and Nevada are not included in the dataset now
This shouldn't be an issue because the 52 crops we are training the models on are rarely if ever planted in these states because of the climate.
I just realized that Alaska hawaii and nevada are still there they just have less than 30k rows. So instead I'm going to try to clean up the column to only contain state names

```python
def replace_region_with_state(location):
    """Function that checks if state name is in LOCATION_DESC if the name contains a state name but is not just
    the state name it converts it into only the state name. For example OREGON SOUTHWEST DOUGLAS becomes OREGON"""
    for state in all_states:
        if state.upper() in location:
            return state.upper()
    return "OTHER"

filtered_df['LOCATION_DESC'] = filtered_df['LOCATION_DESC'].apply(replace_region_with_state)
filtered_df['LOCATION_DESC'].value_counts()
```
