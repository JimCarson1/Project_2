```python
import pandas as pd
from dotenv import load_dotenv
from helper import zip_gz, unzip_gz

pd.set_option('display.max_columns', None)
```

```python
crop_df = pd.read_csv("../data/crop_data.csv")
display(f"Number of rows: {crop_df.shape[0]:,}")
display(crop_df.head())
```

```python
weather_df = pd.read_csv("../data/weather_data.csv")
display(f"Number of rows: {weather_df.shape[0]:,}")
display(weather_df.head())
```

```python
weather_df['state'] = weather_df['state'].str.upper()
```

```python
merged_df = pd.merge(
    crop_df,
    weather_df,
    left_on=['LOCATION_DESC', 'YEAR'],
    right_on=['state', 'year'],
    how='left'
)

merged_df = merged_df.drop(columns=['LOCATION_DESC', 'YEAR'])
merged_df.rename(columns={'COMMODITY_DESC': 'crop', "normalized_value": 'target'}, inplace=True)
merged_df
```

```python
crops = []
for crop in merged_df['crop'].unique():
    crops.append(crop)
```

```python
wheat_df = merged_df.loc[merged_df['crop'] == "WHEAT"]
```

```python
high_target = wheat_df.loc[wheat_df['target'] == 100]
high_target
```

```python
states = []
for state in high_target['state'].unique():
    states.append(state)
len(states)
```

```python
total_finished = 0

# Loop through all crops in full dataset
for crop in crops:
    # Store each speciic crop in df drop uneccesary columns
    df = merged_df.loc[merged_df['crop'] == crop]
    df = df.drop(columns=["crop", "state", "year"])

    # Update user when df is ready to be stored
    print(f"{crop.lower()}_df ready")
    zip_gz(df, f"data_per_crop/{crop.lower()}_df.gz")

    # Update percentage and print it out
    total_finished += df.shape[0]
    percentage = (total_finished / merged_df.shape[0]) * 100
    print(f"Progress: {percentage:.2f}%\n")
```
