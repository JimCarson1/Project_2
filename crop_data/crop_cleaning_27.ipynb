### I want to create a dictionary that will store each crops seperate type of unit.

```python
df_1980 = df.loc[df['YEAR'] >= 1980]
```

```python
crop_units = {}
for crop in tot_crops:
    crop_units[crop] =  df_1980.loc[df_1980['COMMODITY_DESC'] == crop]
```

```python
num_crop = 0
for key, value in crop_units.items():
    num_crop+=1
    print(key)
```

```python
dict_shape = {}
for key, value in crop_units.items():
    dict_shape[key] = value.shape[0]

dict_shape = dict(sorted(dict_shape.items(), key=lambda item: item[1]))

for key, value in dict_shape.items():
    print(f"{key}    {value:,}")
```

```python
years = df['YEAR'].value_counts()

years = dict(sorted(years.items(), key=lambda item: item[0]))

total = 0
for key, value in years.items():
    if int(key) >= 1980:
        print(f"{key}      {value:,}")
        total+=value

```
