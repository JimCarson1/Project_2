```python
import pandas as pd
```

```python
df = pd.read_csv("flu_data/ILINet.csv")
df.reset_index(inplace=True)
df.head(6)
```

```python
df.columns = df.iloc[0]
df = df.drop(0)
df['WEEK'] = df['WEEK'].astype(int)
df.head(6)
```

```python
def week_to_quarter(week):
    if week <= 13:
        return '1'
    elif week <= 26:
        return '2'
    elif week <= 39:
        return '3'
    else:
        return '4'

# Apply the function to create a new column 'QUARTER'
df['QUARTER'] = df['WEEK'].apply(week_to_quarter)
df.head()
```

```python
flu_df = df.drop(columns=['REGION TYPE', 'WEEK', '% WEIGHTED ILI', '%UNWEIGHTED ILI', 'AGE 0-4',	'AGE 25-49',	'AGE 25-64',	'AGE 5-24',	'AGE 50-64',	'AGE 65'])
flu_df['ILITOTAL'] = flu_df['ILITOTAL'].astype(int)
flu_df.head()
```

```python
flu_grouped = flu_df.groupby(['REGION', 'YEAR', 'QUARTER']).agg({
    'ILITOTAL': 'sum',
}).reset_index()
flu_grouped.head()
```

```python
all_states = [
    "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida",
    "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine",
    "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska",
    "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota",
    "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee",
    "Texas", "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"
]
```

```python
pop_df = pd.read_csv("data/state_population.csv")
pop_df.head()
```

```python
pop_df = pop_df.loc[pop_df['NAME'].isin(all_states)]
pop_df = pop_df.drop(columns=['POPESTIMATE042020', 'ESTIMATESBASE2010', 'CENSUS2010POP', 'DIVISION', 'REGION', 'SUMLEV'])
pop_df.head()
```

```python
df_melted = pd.melt(pop_df,
                    id_vars=['NAME'],  # Columns that should remain as is (like 'STATE')
                    value_vars=['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013',
                                'POPESTIMATE2014', 'POPESTIMATE2015', 'POPESTIMATE2016', 'POPESTIMATE2017',
                                'POPESTIMATE2018', 'POPESTIMATE2019', 'POPESTIMATE2020'],  # Columns to unpivot
                    var_name='Year',  # New column for years
                    value_name='Population')  # New column for population estimates

df_melted['Year'] = df_melted['Year'].str.extract('(\d+)')


df_melted.head()
```

```python
merged_df = pd.merge(df_melted, flu_grouped,
                     left_on=['NAME', 'Year'],  # Use 'State_Name' and 'Year' from df1
                     right_on=['REGION', 'YEAR'],  # Use 'STATE' and 'Year_Recorded' from df2
                     how='inner')  # You can choose 'left', 'right', 'outer', or 'inner'
merged_df
```

```python
weather_df = pd.read_csv("data/state_weather_data.csv")

weather_df.head()
```

```python
weather_df['month_day'] = weather_df['date'].apply(lambda x: x[5:])
weather_df['Year'] = weather_df['date'].apply(lambda x: x[:4]).astype(int)

weather_df = weather_df.loc[weather_df['Year'] >= 2010]

all_dates = []

for date in weather_df['month_day'].value_counts().keys():
    all_dates.append(date)

# Create a function to map dates to quarters
def assign_quarter(date):
    month = int(date.split('-')[0])  # Extract month
    if month in [1, 2, 3]:
        return '1'
    elif month in [4, 5, 6]:
        return '2'
    elif month in [7, 8, 9]:
        return '3'
    else:
        return '4'

# Apply the function to assign quarters
weather_df['QUARTER'] = weather_df['month_day'].apply(assign_quarter)

weather_df = weather_df.drop(columns=['lat', 'lon', 'tz', 'date', 'units', 'month_day']).reset_index(drop=True)

weather_df = weather_df.astype({
    'QUARTER': 'int',
})

weather_df = weather_df.groupby(['state', 'Year', 'QUARTER']).mean()

weather_df.head()
```

```python
merged_df = merged_df.astype({
    'QUARTER': 'int',
    'Year': 'int',
})


def calculate_population_density(row):
    state = row['NAME']  # Adjust the column name if it's different
    population = row['Population']

    # Check if the state exists in the dictionary
    if state in state_land_area:
        # Calculate population density
        return round(population / state_land_area[state])
    else:
        # Return NaN or some default value if the state is not found
        return None

# Apply the function to each row in the DataFrame and create a new column 'Population_Density'
merged_df['Population_Density'] = merged_df.apply(calculate_population_density, axis=1)

# Display the DataFrame
merged_df.head()
```

```python
full_df = pd.merge(merged_df, weather_df,
                     left_on=['NAME', 'Year', 'QUARTER'],
                     right_on=['state', 'Year', 'QUARTER'],
                     how='left')

full_df.head()
```
